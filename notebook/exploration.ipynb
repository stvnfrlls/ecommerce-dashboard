```python
import pandas as pd
# import matplotlib.pyplot as plt # charting library

# adjust filename to match what you downloaded
df = pd.read_csv('../data/online_retail_II.csv')

# df.head() # prints first 5 rows of the file
# print(df.shape) # prints the number of rows and columns
# print(df.columns) # prints the column names
# print(df.isnull().sum()) # prints the count of missing values in each column
# print(df.dtypes) # print the data type per column

# shows min/max/average
# print(df['Quantity'].describe())
# print(df['Price'].describe())

# removes rows without Customer ID
# print("Rows before:", df.shape[0])
df = df.dropna(subset=['Customer ID'])
# print("Rows after:", df.shape[0])

# check how many cancellations exist
cancellations = df[df['Invoice'].astype(str).str.startswith('C')] 
# astype(str) - converts column to string
# .str.startswith('C') - check if the column starts with letter C, returns true/false
# df[] - filters the table with rows that only returned true from startswith()
# print("Number of cancellations:", cancellations.shape[0])

# keep everything else as "sales" (non-cancelled orders)
sales = df[~df['Invoice'].astype(str).str.startswith('C')]
# '~' means NOT which means df[] will keep rows that does NOT start with C (Cancelled)
# print("Number of sales rows:", sales.shape[0])

# print("Rows before:", sales.shape[0])
sales = sales[(sales['Quantity'] > 0) & (sales['Price'] > 0)]
# sales[] filters out rows with Quantity and Price with positive value
# print("Rows after:", sales.shape[0])

# print("Before conversion:", sales['InvoiceDate'].dtype)
sales['InvoiceDate'] = pd.to_datetime(sales['InvoiceDate'])
# pd.to_datetime converts value of the column into pandas date time format
# print("After conversion:", sales['InvoiceDate'].dtype)

# sales['Revenue] added new column called Revenue
sales['Revenue'] = sales['Quantity'] * sales['Price']

# check for values of new Revenue Column
# print(sales[['Quantity', 'Price', 'Revenue']].head())

total_revenue = sales['Revenue'].sum()
# sum() adds up all the value of the column
# print("Total Revenue:", total_revenue)

# created new column Year and Month
sales['Year'] = sales['InvoiceDate'].dt.year
sales['Month'] = sales['InvoiceDate'].dt.month

# checked new column
# print(sales[['Month', 'Year', 'Quantity', 'Price', 'Revenue']].head())

year_filter = 2010
sales_year = sales[sales['Year'] == year_filter]
# filters out the rows that matched with year_filter

monthly_revenue = sales_year.groupby('Month')['Revenue'].sum()
# group them per month then calculate for the sum of revenue per each month
# print(monthly_revenue)

# monthly_revenue.plot(kind='bar', figsize=(10, 5)) # plot the values of monthly revenue as bar chart
# plt.title(f'Revenue by Month - {year_filter}') # label for the chart
# plt.xlabel('Month') # label for x axis
# plt.ylabel('Revenue') # label for y axis
# plt.show() # displays chart

sales.to_csv('../data/sales_cleaned.csv', index=False)
# creates a new file for the cleaned data and all added columns above
# print("Saved successfully!")
```
